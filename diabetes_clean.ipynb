```python
##Replacing all zeros with nan
df.replace(0,np.nan,inplace=True)
```

```python
df.head()
```

From the histogram above,BloodPressure, the Glucose,BMI is normally distributed so we replace the Nan values with their means. 
 Insulin,SkinThickness is right skewed so we replace the nan values with mode.

```python
df["Glucose"].fillna(df["Glucose"].mean(),inplace=True)
```

```python
df["BloodPressure"].fillna(df["BloodPressure"].mean(), inplace=True)
df["BMI"].fillna(df["BMI"].mean(),inplace=True)
df["Insulin"].fillna(df["Insulin"].mode(),inplace=True)
df["SkinThickness"].fillna(df["SkinThickness"].mode(),inplace=True)
```

```python
df.replace(np.nan,0,inplace=True)
```
